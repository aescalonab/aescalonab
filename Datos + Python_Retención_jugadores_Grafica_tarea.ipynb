**Retención:** % de jugadores que se vuelven despues de la primera sesión.

```python
import pandas as pd

# Cargar el archivo: recuerda subirlo en la lateral luego de descargarlo en tu computadora, dando click en el foder y luego la primera opción de Cargar.

grupo_datos_procesado = pd.read_csv('candy_crush.csv')

grupo_datos_procesado.head()
```

```python
grupo_datos_procesado.info()
```

```python
grupo_datos_procesado['num_abandonments'] = grupo_datos_procesado['num_attempts']- grupo_datos_procesado['num_success']
```

```python
grupo_datos_procesado.head()
```

```python
# completion_rate : tasa de completado
grupo_datos_procesado['completion_rate'] = grupo_datos_procesado['num_success'] / grupo_datos_procesado['num_attempts']
```

```python
grupo_datos_procesado.head()
```

```python
import matplotlib.pyplot as plt

niveles = [10, 11, 12, 13]
filtro_niveles = grupo_datos_procesado[grupo_datos_procesado['num_success'].isin(niveles)]
tasas_por_nivel = filtro_niveles.groupby('level')['completion_rate'].mean()
tasas_por_nivel = tasas_por_nivel.sort_index()
#para la grafica

plt.figure(figsize=(8, 5))
plt.bar(tasas_por_nivel.index.astype(str), tasas_por_nivel.values, color='skyblue')
plt.title('Tasa de completación por nivel (10 al 13)')
plt.xlabel('Nivel')
plt.ylabel('Tasa de completación promedio')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
